```python
import pandas as pd
import math
L = 14398.16                                                   #rever
cota_mucum = 48
cota_encantado = 31
largura_rio = [153.24,218,130,182.5,185.25,182,180.15,188.59,159.42,226.95]
B = pd.Series(largura_rio).mean()
S = (cota_mucum - cota_encantado) /( L / 1000 ) # em m / Km
S_m = (cota_mucum - cota_encantado) /( L ) # em m / m
n = 0.03
Q_pico = dadosMucum["Vazao"].max()

Q_ref = 0.9* Q_pico


```

```python
h = ((n * Q_ref)/ (B *S_m**0.5))**(3/5)  # h= profundidade area molhada, n = manning , S = declividade , B = largura
print(h)
```

```python

# velocidade e celeridade
V = Q_ref/(h * B)

c = 5/3 * V

```

```python
delta_t = 15*60
# numero de subtrechos
delta_x = c * delta_t / 2 * (1 + ( 1 + 1.5 * Q_ref / (B * S_m * delta_t * c**2))**(1/2))
N = math.ceil(L / delta_x)
delta_x_subtrecho = L / N
N = 20
print(delta_x_subtrecho)
print(N)
```

```python
k = delta_x_subtrecho / c  # delta_x = comprimento do trecho de rio
# parametro X , contrala a atenuação da onda de cheia
X = 0.5 * ( 1 - ( Q_ref/(B * c * S_m * delta_x_subtrecho ) ) )
print(X)
```

```python
C1 = (delta_t - 2 * ( k * X ))/ (2 * k * (1- X) + delta_t)
C2 = (delta_t + 2 * ( k * X ))/ (2 * k * (1- X) + delta_t)
C3 = (2 * k * ( 1 - X ) - delta_t) / ( 2 * k * ( 1 - X ) + delta_t)



print (X ,delta_t / ( 2 * k), (1 - X))
print(C1+C2+C3)
```

```python
import pandas as pd

valores = dadosMucum["Vazao"]

n_colunas = pd.DataFrame({
    'tempo': range(len(valores)),
    'vazao_de_entrada': valores
})


for i in range(1, N + 1):
    col_name = f'Q_saida{i}'
    n_colunas[col_name] = 0

# Definindo os valores no tempo 0
for i in range(1, N + 1):
    col_name = f'Q_saida{i}'
    n_colunas.at[0, col_name] = n_colunas.at[1, 'vazao_de_entrada']


for t in range(1, len(valores)):

    n_colunas.at[t, 'Q_saida1'] = (
        C1 * n_colunas.at[t - 1, 'vazao_de_entrada'] +  # Q_entrada[tempo-1]
        C2 * n_colunas.at[t, 'vazao_de_entrada'] +      # Q_entrada[tempo]
        C3 * n_colunas.at[t - 1, 'Q_saida1']             # Q_saida1[tempo-1]
    )

    # Calculando Q_saida2 até Q_saidaN com base nas vazões anteriores
    for i in range(2, N + 1):
        col_name = f'Q_saida{i}'
        prev_col_name = f'Q_saida{i - 1}'

        # Aplicando a fórmula para Q_saida{i}
        n_colunas.at[t, col_name] = (
            C1 * n_colunas.at[t, prev_col_name] +
            C2 * n_colunas.at[t - 1, prev_col_name] +
            C3 * n_colunas.at[t - 1, col_name]
        )




n_colunas = n_colunas.round().astype(int)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(28, 12))
for column in n_colunas.columns[1:]:
    plt.plot(n_colunas['tempo'], n_colunas[column], label=column)

plt.title('Comparação das Vazões e Saídas')
plt.xlabel('Tempo')
plt.ylabel('Vazão')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Função para o modelo Vazao = a * (NivelAdotado - NivelAdotado_min) ** b
def modelo(x, a, b, nivel_min):
    return a * (x - nivel_min) ** b
dadosEncantado['NivelAdotado']=dadosEncantado['NivelAdotado'] /100
# Removendo valores NaN ou Inf das colunas "Vazao" e "NivelAdotado"
dados_validos = dadosEncantado.dropna(subset=["Vazao", "NivelAdotado"])  # Remove linhas com NaN
dados_validos = dados_validos[np.isfinite(dados_validos["Vazao"])]  # Remove valores Inf na coluna Vazao
dados_validos = dados_validos[np.isfinite(dados_validos["NivelAdotado"])]  # Remove valores Inf na coluna NivelAdotado



# Encontrando o valor mínimo de NivelAdotado
nivel_min = dados_validos["NivelAdotado"].min()

# Ajustando os parâmetros a e b
parametros, covariancia = curve_fit(lambda x, a, b: modelo(x, a, b, nivel_min),
                                    dados_validos["NivelAdotado"], dados_validos["Vazao"])

# Obtendo os parâmetros ajustados
a, b = parametros

# Criando o gráfico
plt.figure(figsize=(28, 12))

# Plotando o gráfico de dispersão dos dados originais
plt.scatter(dados_validos["Vazao"], dados_validos["NivelAdotado"], label="Dados Observados")

# Plotando a curva ajustada
x_fit = np.linspace(dados_validos["NivelAdotado"].min(), dados_validos["NivelAdotado"].max(), 1000)
y_fit = modelo(x_fit, a, b, nivel_min)
plt.plot(y_fit, x_fit, color='red', label=f'Ajuste: Vazao = {a:.2f} * (NivelAdotado - {nivel_min:.2f})^{b:.2f}')

# Adicionando título e rótulos aos eixos
plt.title('Curva chave Encantado')
plt.xlabel('Vazão')
plt.ylabel('Nível Adotado')

# Exibindo a legenda e a grade
plt.legend()
plt.grid()

# Mostrando o gráfico
plt.show()
```
